```python
#QUAL LOJA POSSUI O MAIOR NÚMERO DE RECLAMAÇÕES
# QUAL LOJA POSSUI O MAIOR INDICE DE SOLUÇÃO
# QUAL LOJA QUE MAIS RESPONDE OS CLIENTES
# QUAL LOJA POSSUI A PIOR NOTA DO CONSUMIDOR
# QUAL LOJA POSSUI A MELHOR NOTA DE CONSUMIDOR
# QUAL LOJA AS PESSOAS MAIS VOLTARIAM FAZER NEGOCIO DEPOIS DE UMA RECLAMAÇÃO
# QUAL LOJA POSSUI A PIOR REPUTAÇÃO
```

**LOJA QUE POSSUI O MAIOR NUMERO DE RECLAMAÇÕES**

```python
df[['Nome_Da_Loja']][df['Total_de_reclamaçoes']==df['Total_de_reclamaçoes'].max()]
```

**LOJA QUE POSSUI O MENOR NUMERO DE RECLAMAÇÕES**

```python
df[['Nome_Da_Loja']][df['Total_de_reclamaçoes']==df['Total_de_reclamaçoes'].min()]
```

**Loja em que os clientes mais voltariam a fazer compras** *italicized text*

```python
df[['Nome_Da_Loja']][df['Voltariam_a_fazer_negogicio']==df['Voltariam_a_fazer_negogicio'].max()]
```

**Loja em que os clientes menos voltariam a fazer compras**

```python
df[['Nome_Da_Loja']][df['Voltariam_a_fazer_negogicio']==df['Voltariam_a_fazer_negogicio'].min()]
```

**Loja que possui a melhor nota do consumidor**

```python
df[['Nome_Da_Loja']][df['Nota_do_consumidor']==df['Nota_do_consumidor'].max()]
```

**Loja que possui a pior  nota do consumidor**

```python
df[['Nome_Da_Loja']][df['Nota_do_consumidor']==df['Nota_do_consumidor'].min()]
```

**Loja que possui a maior reputação**

```python
df[['Nome_Da_Loja']][df['Reputação']==df['Reputação'].max()]
```

**Loja que possui a menor reputação**

```python
df[['Nome_Da_Loja']][df['Reputação']==df['Reputação'].min()]
```

**LOJAS QUE POSSUEM UMA REPUTAÇÃO MAIOR QUE 6**

```python
lojas = df[df['Reputação']>6]
lojas.plot(kind='bar', x='Nome_Da_Loja', y='Reputação')
```

```python
lojas2 = df[df['Reputação']>8]
lojas2.plot(kind='bar', x='Nome_Da_Loja', y='Reputação')
```
